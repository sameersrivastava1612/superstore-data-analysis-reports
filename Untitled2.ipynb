```python
import pandas as pd
import numpy as np

df.columns = [c.strip().lower().replace(" ", "_").replace("-", "_").replace("/", "_") for c in df.columns]


df.columns
```

```python

df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])


df['order_year'] = df['order_date'].dt.year
df['order_month'] = df['order_date'].dt.to_period('M').astype(str)
df['shipping_days'] = (df['ship_date'] - df['order_date']).dt.days


df.drop_duplicates(inplace=True)


df.info()
df.head()
```

```python
total_sales = df['sales'].sum()
total_profit = df['profit'].sum()
total_orders = df['order_id'].nunique()
avg_order_value = total_sales / total_orders
profit_margin = total_profit / total_sales

kpis = {
    "Total Sales": total_sales,
    "Total Profit": total_profit,
    "Total Orders": total_orders,
    "Average Order Value": avg_order_value,
    "Profit Margin": profit_margin
}

kpis
```

```python
import matplotlib.pyplot as plt

df_month = df.groupby('order_month')['sales'].sum().reset_index()
df_month['order_month'] = pd.to_datetime(df_month['order_month'])

plt.figure(figsize=(12,6))
plt.plot(df_month['order_month'], df_month['sales'])
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns

region_sales = df.groupby('region')['sales'].sum().reset_index()

sns.barplot(data=region_sales, x='region', y='sales')
plt.title("Sales by Region")
plt.show()
```

```python
cat = df.groupby('category')['sales'].sum().reset_index()
subcat = df.groupby('sub_category')['sales'].sum().reset_index()

print("Category Sales:")
display(cat)

print("Sub-Category Sales:")
display(subcat)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.heatmap(df[['sales','profit','discount','quantity','shipping_days']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
top_products = df.groupby('product_name')['sales'].sum().sort_values(ascending=False).head(10)
top_products
```

```python
top_products.plot(kind='bar', figsize=(12,6), title="Top 10 Best-Selling Products")
plt.ylabel("Sales")
plt.show()
```

```python
segment_profit = df.groupby('segment')['profit'].sum().reset_index()

sns.barplot(data=segment_profit, x='segment', y='profit')
plt.title("Profit by Segment")
plt.show()
```

```python
loss_products = df[df['profit'] < 0].groupby('product_name')['profit'].sum().sort_values().head(10)
loss_products
```

```python
ship_perf = df.groupby('ship_mode')['shipping_days'].mean().reset_index()

sns.barplot(data=ship_perf, x='ship_mode', y='shipping_days')
plt.title("Average Shipping Days by Ship Mode")
plt.show()
```
